## SQLiteを使う際の基本的なDB処理の流れ
1. DBへのコネクションを確立する
2. コネクションオブジェクトからデータ操作用のカーソルオブジェクトを取得する
3. SQLを定義する
4. SQLを実行する
5. 必要があればコミットする
6. コネクションを閉じる

```python
import sqlite3
```

```python
# 1. DBへのコネクションを確立する
conn = sqlite3.connect('weather_data.db')
```

```python
# 2. コネクションオブジェクトからデータ操作用のカーソルオブジェクトを取得する
cursor = conn.cursor()
```

```python
# 3. SQLを定義する
create_table_sql = '''
CREATE TABLE weather_data (
    date TEXT,
    localatm INT,
    seaatm INT,
    allamp INT,
    '60amp' INT,
    '10amp' INT,
    avgtep INT,
    maxtep INT,
    mintep INT,
    avghum INTEGER,
    minhum INTEGER,
    avgws INT,
    maxws INT,
    maxwsd TEXT,
    maxwsm INT,
    maxwsmd TEXT,
    sunh INT,
    snow INT,
    maxsnow INT,
    amweather TEXT,
    pmweather TEXT
);
'''
```

```python
cursor.execute(create_table_sql)

# 5. コミットする
conn.commit()

# 6. コネクションを閉じる
conn.close()
```
